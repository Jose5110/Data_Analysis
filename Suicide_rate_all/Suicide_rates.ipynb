```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('suiciderateall.csv')
df_male = pd.read_csv('suicideratemale.csv')
df_female = pd.read_csv('suicideratefemale.csv')
```

## We begin the analysis with the dataset for the overall suicide rate per country for the period from 2000 to 2019

```python
# Initial visualization of the overall dataset for reference
print('Dataframe shape: '+f'{df.shape}')
df.head()
```

### Cleaning the Dfs

```python
# Overall
df['Country'] = df['Country'].str.replace('*','').str.strip()

# Males
df_male['Country'] = df_male['Country'].str.replace('*','').str.strip()

# Females
df_female['Country'] = df_female['Country'].str.replace('*','').str.strip()
```

## Suicide rate comparison by 5 year periods from 2000 to 2019

```python
plt.style.use('fivethirtyeight')
# Overall
plt.figure(figsize=(15,5))
labels = np.arange(1,6)
x = np.arange(5)

plt.plot(x,df.describe().iloc[1, 0:5], label='2000-2004', marker='o', markersize=10)
plt.plot(x,df.describe().iloc[1, 5:10], label='2005-2009', marker='o', markersize=10)
plt.plot(x,df.describe().iloc[1, 10:15], label='2010-2014', marker='s', markersize=10)
plt.plot(x,df.describe().iloc[1, 15:20], label='2015-2019', marker='s', markersize=10)

plt.xticks(x, labels)
plt.xlabel('Years (study period)')
plt.ylabel('Suicide rate average')
plt.title('Worldwide overall suicide rate by 5 year periods (2000-2019)')
plt.legend(bbox_to_anchor=(1, 1), loc="upper left")
plt.show()

# Male
plt.figure(figsize=(15,5))
labels = np.arange(1,6)
x = np.arange(5)

plt.plot(x,df_male.describe().iloc[1, 0:5], label='2000-2004', marker='o', markersize=10)
plt.plot(x,df_male.describe().iloc[1, 5:10], label='2005-2009', marker='o', markersize=10)
plt.plot(x,df_male.describe().iloc[1, 10:15], label='2010-2014', marker='s', markersize=10)
plt.plot(x,df_male.describe().iloc[1, 15:20], label='2015-2019', marker='s', markersize=10)

plt.xticks(x, labels)
plt.xlabel('Years (study period)')
plt.ylabel('Suicide rate average')
plt.title('Worldwide male suicide rate by 5 year periods (2000-2019)')
plt.legend(bbox_to_anchor=(1, 1), loc="upper left")
plt.show()

# Female
plt.figure(figsize=(15,5))
labels = np.arange(1,6)
x = np.arange(5)

plt.plot(x,df_female.describe().iloc[1, 0:5], label='2000-2004', marker='o', markersize=10)
plt.plot(x,df_female.describe().iloc[1, 5:10], label='2005-2009', marker='o', markersize=10)
plt.plot(x,df_female.describe().iloc[1, 10:15], label='2010-2014', marker='s', markersize=10)
plt.plot(x,df_female.describe().iloc[1, 15:20], label='2015-2019', marker='s', markersize=10)

plt.xticks(x, labels)
plt.xlabel('Years (study period)')
plt.ylabel('Suicide rate average')
plt.title('Worldwide female suicide rate by 5 year periods (2000-2019)')
plt.legend(bbox_to_anchor=(1, 1), loc="upper left")
plt.show()

# Male vs Female
plt.figure(figsize=(15,5))
labels = np.arange(2000,2020)
x = np.arange(20)

plt.plot(x,df_male.describe().iloc[1], label='Male', marker='o', markersize=10)
plt.plot(x,df_female.describe().iloc[1], label='Female', marker='s', markersize=10)

plt.xticks(x, labels)
plt.xlabel('Year')
plt.ylabel('Suicide rate average')
plt.title('Male vs Female suicide rate comparison')
plt.legend(bbox_to_anchor=(1, 1), loc="upper left")
plt.show()
```

As we can see there is a downwards trend in the overall suicide rates in the world regarding the data supplied in this
dataset, the only instance of this not been a continous decrease in the whole 20 years is in the 2005-2009 period, especifically the year 2008.

The gender divided charts present us with the same overlook, a continuous drecrease in all the periods with the exception been in the year 2008 for both genders. Another insight granted is the large difference between male and female suicide rate, been the male rates significantly larger the female ones.

## Average rate per country for the 2000-2019 period

```python
df['Average rate 2000-2019'] = df.mean(axis=1)
rate_per_country = df[['Country', 'Average rate 2000-2019']]
rate_per_country
```

## Top 10 countries by suicide rate (Higher is worst)

```python
top_10_high = rate_per_country.sort_values(by='Average rate 2000-2019', ascending=False)[0:10].reset_index(drop=True)
top_10_high
```

```python
fig, ax = plt.subplots(figsize= (10,5))
x_value = top_10_high['Average rate 2000-2019']
y_value = top_10_high['Country']
y_pos = np.arange(len(y_value))
plt.barh(y_pos, x_value, edgecolor='black', linewidth=1)
plt.yticks(y_pos, y_value)
plt.gca().invert_yaxis()
ax.set(title='Top 10 countries by suicide rate (Higher is worst)', ylabel= 'Country', xlabel= 'Suicide rate');
plt.show()
```

## Top 10 countries by lower suicide rate (Lower is better)

```python
top_10_low = rate_per_country.sort_values(by='Average rate 2000-2019')[0:10].reset_index(drop=True)
top_10_low
```

```python
fig, ax = plt.subplots(figsize= (10,5))
x_value = top_10_low['Average rate 2000-2019']
y_value = top_10_low['Country']
y_pos = np.arange(len(y_value))
plt.barh(y_pos, x_value, edgecolor='black', linewidth=1)
plt.yticks(y_pos, y_value)
plt.yticks(rotation=25)
plt.gca().invert_yaxis()
ax.set(title='Top 10 countries by lower suicide rate (Lower is better)', ylabel= 'Country', xlabel= 'Suicide rate');
plt.show()
```

## Year with the highest suicide rate for each country

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
max_country = pd.DataFrame({ 'Country': df['Country'],'Max year avg': df.max(axis=1),'Year': df.iloc[:,1:].idxmax(axis=1)})
max_country
```

## Year with the lowest suicide rate for each country

```python
min_country = pd.DataFrame({ 'Country': df['Country'],'Min year avg': df.min(axis=1),'Year': df.iloc[:,1:].idxmin(axis=1)})
min_country
```

## Worst year?

```python
year_max_count = max_country.groupby('Year').agg({'Country':'count'})
year_max_count.rename(columns={'Country':'Countries'}, inplace=True)
year_max_count.reset_index(inplace=True)
print('Worst year by max suicide rates per country ')
year_max_count
```

```python
plt.figure(figsize=(10,5))
labels = np.arange(2000,2020)
x = np.arange(20)

plt.plot(year_max_count.Year, year_max_count.Countries, label='Countries', marker='o', markersize=10)
plt.xticks(x, labels)
plt.xticks(rotation=60)
plt.xlabel('Year')
plt.ylabel('Countries')
plt.title('Worst year by max suicide rates per country')
plt.show()
```

As we can see the year 2000 was by far the worst, there's a lot of factors that could have influenced this and the fact that the year itself is the begining of the new millenium is an important one. 

## ...and the best?

```python
year_min_count = min_country.groupby('Year').agg({'Country':'count'})
year_min_count.rename(columns={'Country':'Countries'}, inplace=True)
year_min_count.reset_index(inplace=True)
print('Best year by min suicide rates per country ')
year_min_count
```

```python
plt.figure(figsize=(10,5))
labels = np.arange(2000,2020)
x = np.arange(20)

plt.plot(year_min_count.Year, year_min_count.Countries, label='Countries', marker='o', markersize=10)
plt.xticks(x, labels)
plt.xticks(rotation=60)
plt.xlabel('Year')
plt.ylabel('Countries')
plt.title('Best year by min suicide rates per country')
plt.show()
```

As we can see, in agreement with the analysis of suicide rates comparison by 5 year periods, the continual decrease of suicide rates through the years reaches its minimun at the end of the study period, giving us 2019 as the best year regarding suicide rates worldwide.

A point to clarify on this last analysis is the difference between a minimum in the overall suicide rate average and the amount of countries that reach a minimun rate in a given year, countries with a disproportionate rate of suicides could very well experience a sudden decline in this rates bringing the overall average down while at the same time several countries with small rates see an increase in their indexes, in this case the correct combination could present us with a decrease of the world average but an increase in the rates all across the table.
